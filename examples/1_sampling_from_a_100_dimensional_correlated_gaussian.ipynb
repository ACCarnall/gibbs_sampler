Sampling from a 100 Dimensional correlated Gaussian
===========================================

This example demonstrates the ability of the HAGGiS sampler to efficiently sample higher-dimensional posteriors.

First we will set up a likelihood function for a correlated 100-D Gaussian:

```python
import numpy as np
import haggis
import corner

import matplotlib.pyplot as plt
%matplotlib inline

# Seed the random number generator
np.random.seed(0)

# Set up the covariance matrix
n_dim = 10
cov = np.identity(n_dim)
#cov[cov == 0] = 0.9
        
cov_inv = np.linalg.inv(cov)

def lnlike(param, ndim=0, nparam=0):
    """ Log-likelihood for a corellated Gaussian. """
    return -0.5*np.dot(param, np.dot(cov_inv, param))
```

The sampler is set up to take the same inputs as common nested sampling algorithms. Therefore we must also define the prior transform function which maps numbers from a unit cube onto our prior volume. We'll use a uniform prior from -5 to 5 in each parameter.

```python
def prior_transform(cube, ndim=0, nparam=0):
    return 20.*cube - 10.
        
```

Now we can set up and run the sampler:

```python
n_steps = 5000

sampler = haggis.gibbs_sampler(lnlike, prior_transform, n_dim, cdf_accuracy=0.01)
    
sampler.run(n_steps, verbose=True)
```

```python
for i in range(3):
    plt.figure(figsize=(10, 3))
    plt.plot(np.arange(n_steps+1), sampler.samples[:, i])
    plt.show()
```

```python
corner.corner(sampler.samples)
plt.show()
```
